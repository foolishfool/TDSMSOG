```python
#THIS CODE ONLY DO THE WHOLE DATA TRAINING, NO CROSS_VALIDATION AND FEATURE SELECTION
import pandas as pd
import numpy as np
import dataset_read
import warnings
import experiment
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
dataread = dataset_read.DATAREAD()
```

```python
csv = pd.read_csv("Mobile Price Classification/train.csv")
csv_original_encode = csv

#dataread.hash_encoding(csv_original_encode,["blue","dual_sim","fc","four_g","mobile_wt","n_cores","sc_h","sc_w","talk_time","three_g","touch_screen","wifi"])
dataread.label_encoding(csv,"blue")
dataread.label_encoding(csv,"dual_sim")
dataread.label_encoding(csv,"fc")
dataread.label_encoding(csv,"four_g")
dataread.label_encoding(csv,"mobile_wt")
dataread.label_encoding(csv,"n_cores")
dataread.label_encoding(csv,"sc_h")
dataread.label_encoding(csv,"sc_w")
dataread.label_encoding(csv,"talk_time")
dataread.label_encoding(csv,"three_g")
dataread.label_encoding(csv,"touch_screen")
dataread.label_encoding(csv,"wifi")

csv_test = csv.sample(int(csv.shape[0]*0.3))
csv_training = csv.drop(csv_test.index,axis = 0)

dataread.initializedataset(csv_training,csv_test,"price_range")
```
