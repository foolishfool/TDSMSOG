```python
#THIS CODE ONLY DO THE WHOLE DATA TRAINING, NO CROSS_VALIDATION AND FEATURE SELECTION
import pandas as pd
import numpy as np
import dataset_read
import warnings
import experiment
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
dataread = dataset_read.DATAREAD()
```

```python


csv = pd.read_csv("student_dropout_rwanda_education.csv")


dataread.label_encoding(csv,"gender")
dataread.label_encoding(csv,"orphans")
dataread.label_encoding(csv,"paying_school_feeding")
dataread.label_encoding(csv,"parental_education")



csv_test = csv.sample(int(csv.shape[0]*0.3))
csv_training = csv.drop(csv_test.index,axis = 0)

dataread.initializedataset(csv,csv_training,csv_test,"dropout")
```

```python
import experiment
class_num= 3
dim_num = 10
scope_num = 50
experiment = experiment.Experiment()
experiment.Ttest(dataread,class_num, dim_num,scope_num)
```
